```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv("train.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.sample()
```

```python
plt.imshow(df.iloc[39098,1:].values.reshape(28,28))
```

```python
x=df.iloc[:,1:]
y=df.iloc[:,0]
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.neighbors import KNeighborsClassifier
knn=KNeighborsClassifier()
```

```python
knn.fit(x_train,y_train)
```

```python
import time
start=time.time()
y_predict=knn.predict(x_test)
print(time.time()-start)
```

```python
from sklearn.metrics import accuracy_score
y_score=accuracy_score(y_predict,y_test)
```

```python
y_score*100
```

```python
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

```python
from sklearn.decomposition import PCA
```

```python
pca=PCA(n_components=300) #number of principal components
```

```python
x_train_trf=pca.fit_transform(x_train)
x_test_trf=pca.transform(x_test)
```

```python
x_train_trf.shape
```

```python
knn=KNeighborsClassifier()
```

```python
knn.fit(x_train_trf,y_train)
```

```python
from sklearn.metrics import accuracy_score
y_pred=knn.predict(x_test_trf)
accuracy_score(y_pred,y_test)
```

```python
from sklearn.decomposition import PCA
from sklearn.metrics import accuracy_score
for i in range (1,784,50):
    pca=PCA(n_components=i) #number of principal components
    x_train_trf=pca.fit_transform(x_train)
    x_test_trf=pca.transform(x_test)
    
    knn=KNeighborsClassifier()
    knn.fit(x_train_trf,y_train)
    
    y_pred=knn.predict(x_test_trf)
    print(accuracy_score(y_pred,y_test))
```

```python
pca=PCA(n_components=3) #number of principal components
x_train_trf=pca.fit_transform(x_train)
x_test_trf=pca.transform(x_test)
    
knn=KNeighborsClassifier()
knn.fit(x_train_trf,y_train)
    
y_pred=knn.predict(x_test_trf)
print(accuracy_score(y_pred,y_test))
```
