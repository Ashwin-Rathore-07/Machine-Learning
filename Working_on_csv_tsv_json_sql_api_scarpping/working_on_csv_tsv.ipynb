```python
#csv file from url (i have used github)
import requests
from io import StringIO
# Raw GitHub CSV URL (not the normal GitHub page link)
url = "https://raw.githubusercontent.com/MainakRepositor/Datasets/master/Air%20Quality/real_2016_air.csv"
# Add User-Agent header to avoid request blocking
headers = {
    "User-Agent": "Mozilla/5.0 (Macintosh; Intel Mac OS X 10_14) "
                  "AppleWebKit/537.36 (KHTML, like Gecko) "
                  "Chrome/66.0.3359.181 Safari/537.36"
}
# Fetch CSV file
req = requests.get(url, headers=headers)
data = StringIO(req.text)

df = pd.read_csv(data)
print(df.head())
```
