```python
import requests
import pandas as pd

api_key = "2cbf34b5"
url = f"http://www.omdbapi.com/?s=Avengers&apikey={api_key}"

response = requests.get(url)
data = response.json()

# "Search" contains a list of movies
movies_list = data.get("Search", [])

df = pd.DataFrame(movies_list, columns=['Title', 'Year', 'imdbID', 'Type'])
print(df)
```
